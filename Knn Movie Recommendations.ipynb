```
def read_item_names():
  file_name = get_dataset_dir() + '/ml-100k/ml-100k/u.item'
  rid_to_name = {}
  name_to_rid = {}
  with io.open(file_name,'r',encoding= 'ISO-8859-1') as f:
     for line in f:
       line= line.split('|')
       rid_to_name[line[0]]= line[1]
       name_to_rid[line[0]]= line[0]

  return rid_to_name, name_to_rid
```

```
# first , first train the algorithm to compute the similarities between items
data = Dataset.load_builtin('ml-100k') # data loaded
trainset = data.build_full_trainset() # full data trained for similarities between items on KNNBaseline 
sim_options = {'name': 'pearson_baseline','user_based': False}
algo = KNNBaseline(sim_options==sim_options)
algo.fit(trainset)
```

```
# Read the mappings raw id <-> movie name
rid_tp_name, name_to_rid = read_item_names()

# Retrive inner id of the movie Toy Story
toy_story_raw_id = name_to_rid['Toy Story (1995)']
toy_story_inner_id = algo.trainset(toy_story_inner_id, k=10)

# Convert inner ids of the nearest neighbours of Toy Story
toy_story_neighbours = (algo.trainset.to_raw_iid(inner_id)
                        for inner_id in toy_story_neighbours)

toy_story_neighbours = ( rid_to_name[rid]
                        for rid in toy_story_neighbours)
for movie in toy_story_neighbours :
  print(movie)                            
```
